```python
#conf_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
conf_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
confirmed_df = pd.read_csv(conf_url)

#death_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv"
death_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
deaths_df = pd.read_csv(death_url)

#recov_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv"
#recoveries_df = pd.read_csv(recov_url)
```

```python
nordic = ["Sweden", "Norway", "Denmark"]
nordic_conf_df = confirmed_df[confirmed_df["Country/Region"].isin(nordic)]
nordic_deaths_df = deaths_df[deaths_df["Country/Region"].isin(nordic)]
#nordic_recoveries_df = recoveries_df[recoveries_df["Country/Region"].isin(nordic)]
```

```python
nordic_conf_df = nordic_conf_df.drop(["Province/State","Lat", "Long"], axis=1).groupby("Country/Region").sum()
nordic_deaths_df = nordic_deaths_df.drop(["Province/State","Lat", "Long"], axis=1).groupby("Country/Region").sum()
#nordic_recoveries_df = nordic_recoveries_df.drop(["Province/State","Lat", "Long"], axis=1).groupby("Country/Region").sum()
```

```python
cols = confirmed_df.keys()

confirmed = confirmed_df.loc[:, cols[4]:cols[-1]]
deaths = deaths_df.loc[:, cols[4]:cols[-1]]
#recoveries = recoveries_df.loc[:, cols[4]:cols[-1]]
```

```python
nordic_conf_df["3/10/20"]["Norway"]
```

```python
dates = deaths.keys()
world_cases = []
total_deaths = [] 
mortality_rate = []
#total_recovered = [] 
sweden_cases = []
sweden_deaths = []
#sweden_recov = []
norway_cases = []
norway_deaths = []
#norway_recov = []
denmark_cases = []
denmark_deaths = []
#denmark_recov = []

for i in dates:
    confirmed_sum = confirmed[i].sum()
    sweden_sum = nordic_conf_df[i]["Sweden"]
    norway_sum = nordic_conf_df[i]["Norway"]
    denmark_sum = nordic_conf_df[i]["Denmark"]
    death_sum = deaths[i].sum()
    sweden_death_sum = nordic_deaths_df[i]["Sweden"]
    norway_death_sum = nordic_deaths_df[i]["Norway"]
    denmark_death_sum = nordic_deaths_df[i]["Denmark"]
    #recovered_sum = recoveries[i].sum()
    #sweden_recov_sum = nordic_recoveries_df[i]["Sweden"]
    #norway_recov_sum = nordic_recoveries_df[i]["Norway"]
    #denmark_recov_sum = nordic_recoveries_df[i]["Denmark"]
                                               
    sweden_cases.append(sweden_sum)
    norway_cases.append(norway_sum)
    denmark_cases.append(denmark_sum)
    world_cases.append(confirmed_sum)
    total_deaths.append(death_sum)
    sweden_deaths.append(sweden_death_sum)
    norway_deaths.append(norway_death_sum)
    denmark_deaths.append(denmark_death_sum)
    mortality_rate.append(death_sum/confirmed_sum)
    #total_recovered.append(recovered_sum)
    #sweden_recov.append(sweden_recov_sum)
    #norway_recov.append(norway_recov_sum)
    #denmark_recov.append(denmark_recov_sum)
```

```python
X_train_confirmed, X_test_confirmed, y_train_confirmed, y_test_confirmed = train_test_split(days_since_1_22, world_cases_arr, test_size=0.1, shuffle=False) 
```

```python
kernel = ['linear', 'poly', 'sigmoid', 'rbf']
c = [0.01, 0.1, 1, 10]
gamma = [0.01, 0.1, 1]
epsilon = [0.01, 0.1, 1]
shrinking = [True, False]
svm_grid = {'kernel': kernel, 'C': c, 'gamma' : gamma, 'epsilon': epsilon, 'shrinking' : shrinking}

svm = SVR()
svm_search = RandomizedSearchCV(svm, svm_grid, scoring='neg_mean_squared_error', cv=3, return_train_score=True, n_jobs=-1, n_iter=40, verbose=1)
svm_search.fit(X_train_confirmed, y_train_confirmed)
```

```python
svm_search.best_params_
```

```python
svm_confirmed = svm_search.best_estimator_
svm_pred = svm_confirmed.predict(future_forcast)
```

```python
print(svm_confirmed.predict(X_test_confirmed))
print(y_test_confirmed)
print(type(y_test_confirmed))
```

```python
# check against testing data
svm_test_pred = svm_confirmed.predict(X_test_confirmed)
sns.lineplot(data=svm_test_pred, label="svm")
sns.lineplot(data=y_test_confirmed[:,0], label="y_test")
print('MAE:', mean_absolute_error(svm_test_pred, y_test_confirmed))
print('MSE:',mean_squared_error(svm_test_pred, y_test_confirmed))
```

```python
linear_model = LinearRegression(normalize=True)
linear_model.fit(X_train_confirmed, y_train_confirmed)
test_linear_pred = linear_model.predict(X_test_confirmed)
linear_pred = linear_model.predict(future_forcast)
print('MAE:', mean_absolute_error(test_linear_pred, y_test_confirmed))
print('MSE:',mean_squared_error(test_linear_pred, y_test_confirmed))
```

```python
linear_model.coef_
```

```python
sns.lineplot(data=y_test_confirmed[:,0], label="y_test")
sns.lineplot(data=test_linear_pred[:,0], label="test_linear_pred")
```

```python
tol = [1e-4, 1e-3, 1e-2]
alpha_1 = [1e-7, 1e-6, 1e-5, 1e-4]
alpha_2 = [1e-7, 1e-6, 1e-5, 1e-4]
lambda_1 = [1e-7, 1e-6, 1e-5, 1e-4]
lambda_2 = [1e-7, 1e-6, 1e-5, 1e-4]

bayesian_grid = {'tol': tol, 'alpha_1': alpha_1, 'alpha_2' : alpha_2, 'lambda_1': lambda_1, 'lambda_2' : lambda_2}

bayesian = BayesianRidge()
bayesian_search = RandomizedSearchCV(bayesian, bayesian_grid, scoring='neg_mean_squared_error', cv=3, return_train_score=True, n_jobs=-1, n_iter=40, verbose=1)
bayesian_search.fit(X_train_confirmed, y_train_confirmed)
```

```python
bayesian_search.best_params_
```

```python
bayesian_confirmed = bayesian_search.best_estimator_
test_bayesian_pred = bayesian_confirmed.predict(X_test_confirmed)
bayesian_pred = bayesian_confirmed.predict(future_forcast)
print('MAE:', mean_absolute_error(test_bayesian_pred, y_test_confirmed))
print('MSE:',mean_squared_error(test_bayesian_pred, y_test_confirmed))
```

```python
latest_confirmed = confirmed_df[dates[-1]]
latest_deaths = deaths_df[dates[-1]]
#latest_recoveries = recoveries_df[dates[-1]]
```

```python
unique_countries =  list(confirmed_df['Country/Region'].unique())
```

```python
country_confirmed_cases = []
no_cases = []
for i in unique_countries:
    cases = latest_confirmed[confirmed_df['Country/Region']==i].sum()
    if cases > 0:
        country_confirmed_cases.append(cases)
    else:
        no_cases.append(i)
        
for i in no_cases:
    unique_countries.remove(i)
```

```python
# number of cases per country/region
for i in range(len(unique_countries)):
    print(f'{unique_countries[i]}: {country_confirmed_cases[i]} cases')
```

```python
unique_provinces =  list(confirmed_df['Province/State'].unique())
```

```python
province_confirmed_cases = []
no_cases = [] 
for i in unique_provinces:
    cases = latest_confirmed[confirmed_df['Province/State']==i].sum()
    if cases > 0:
        province_confirmed_cases.append(cases)
    else:
        no_cases.append(i)
 
# remove areas with no confirmed cases
for i in no_cases:
    unique_provinces.remove(i)
```

```python
# number of cases per province/state/city

for i in range(len(unique_provinces)):
    print(f'{unique_provinces[i]}: {province_confirmed_cases[i]} cases')
```

```python
nan_indices = [] 

# handle nan if there is any, it is usually a float: float('nan')

for i in range(len(unique_provinces)):
    if type(unique_provinces[i]) == float:
        nan_indices.append(i)

unique_provinces = list(unique_provinces)
province_confirmed_cases = list(province_confirmed_cases)

for i in nan_indices:
    unique_provinces.pop(i)
    province_confirmed_cases.pop(i)
```
